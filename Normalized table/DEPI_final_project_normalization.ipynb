```python
import pandas as pd

# Load merged file
df = pd.read_csv(r"C:\Users\gehad\Desktop\DEPI\Final project\merged_final.csv")

# Check basic info
print("Data shape:", df.shape)
df.head()
```

```python
#Create Lookup (Dimension) Tables
#Lookup tables store unique items (no duplication).

# Country table
countries = df[['Country']].drop_duplicates().reset_index(drop=True)
countries['country_id'] = countries.index + 1

# Disease table
diseases = df[['Disease', 'Disease Category']].drop_duplicates().reset_index(drop=True)
diseases['disease_id'] = diseases.index + 1

# Food table
foods = df[['food_name', 'description', 'url']].drop_duplicates().reset_index(drop=True)
foods['food_id'] = foods.index + 1
```

```python
df.columns
```

```python
#Add IDs to the main dataset
#We link back the IDs from lookup tables into the merged dataframe.
df = df.merge(countries, on='Country', how='left')
df = df.merge(diseases, on='Disease', how='left')
df = df.merge(foods, on='food_name', how='left')
```

```python
#Create the Normalized Fact Tables

# Air Quality Table
tbl_air_quality = df[['country_id', 'Year', 'Date', 'PM2.5 (µg/m³)', 'PM10 (µg/m³)', 'NO2 (ppb)', 'AQI']].drop_duplicates()

# Health Table
tbl_health = df[['Country', 'Disease', 'Year','Prevalence', 'Incidence', 'Mortality', 'Population Affected','Healthcare Access (%)', 'Doctors per 1000', 'Hospital Beds per 1000','Recovery Rate (%)', 'Improvement in 5 Years (%)', 'Urbanization Rate (%)']].drop_duplicates()

# Food Consumption Table
tbl_food_consumption = df[['country_id', 'Year','food_id', 'description']].drop_duplicates()
```

```python
#Save the normalized tables as CSV
countries.to_csv(r"C:\Users\gehad\Desktop\DEPI\Final project\tbl_country.csv", index=False)
diseases.to_csv(r"C:\Users\gehad\Desktop\DEPI\Final project\tbl_disease.csv", index=False)
foods.to_csv(r"C:\Users\gehad\Desktop\DEPI\Final project\tbl_food.csv", index=False)

tbl_air_quality.to_csv(r"C:\Users\gehad\Desktop\DEPI\Final project\tbl_air_quality.csv", index=False)
tbl_health.to_csv(r"C:\Users\gehad\Desktop\DEPI\Final project\tbl_health.csv", index=False)
tbl_food_consumption.to_csv(r"C:\Users\gehad\Desktop\DEPI\Final project\tbl_food_consumption.csv", index=False)
```
